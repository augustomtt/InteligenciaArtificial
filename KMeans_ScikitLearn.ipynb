# Implementaci√≥n de la actividad 2. Clustering: KMeans





# Imports necesarios

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import requests
```

# Ploteo de los puntos iniciales


```python
myfile = requests.get('https://pastebin.com/raw/tR99Fa3f')
datos = myfile.text
datos = datos.split()
puntos = []
for i in range(len(datos)//2):
    aux = [float(datos[i*2]),float(datos[2*i+1])]
    puntos.append(aux)

x = np.array([])
y = np.array([])
for i in puntos:
    x = np.append(x, i[0])
    y = np.append(y, i[1])
plt.scatter(x,y)
plt.show()
```

# Implementaci√≥n de KMeans usando scikit-learn y ploteo


```python
km = KMeans(3)  # Cantidad de clusters!
etiquetasPuntos = km.fit_predict(puntos)
grupos = np.unique(etiquetasPuntos)
centros = km.cluster_centers_

colores = ["red", "green", "blue", "magenta", "hotpink", "peru", "yellow", "cyan"]

for gp in grupos:
    x = np.array([])
    y = np.array([])
    for i in range(len(puntos)):
        if etiquetasPuntos[i] == gp:
            x = np.append(x, puntos[i][0])
            y = np.append(y, puntos[i][1])
        plt.scatter(x,y,color = colores[gp])
        #plt.scatter(datos[:,0],datos[:,1],c=etiquetas,cmap='rainbow')
plt.scatter(centros[:,0],centros[:,1],s=15,color="black",marker="x")

plt.show()
```
